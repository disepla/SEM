```python
import numpy as np
from scipy.io import loadmat
import beachpy as bp
import matplotlib.pyplot as plt
```

```python
# Beach Data
data = loadmat('matlab_file.mat')
```

```python
#Ribeira
#tanP 0.03 | xP -0.5 | zB 3.1 | tanB 0.05  

RI = data['Ribeira'][:,0]

Vol_RI = []
for i in RI:
    if np.isnan(i):
        Vol_RI.append(0)
    else:
        A = bp.BeachProfile(x_beachface_toe=i, platform_slope =0.03, y_rocky_coastline= -0.5,
                            y_sandy_coastline= 3.1, beachface_slope = 0.05, berm_slope = 0, upper_bound = 2,
                            lower_bound = -4, right_bound = 80)
        plt.figure()
        A.plot()
        Vol_RI.append(A.volume())
        
print('Volume_Ribeira =')
print(Vol_RI)
```
